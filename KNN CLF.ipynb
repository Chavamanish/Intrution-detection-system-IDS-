```python
import pandas as pd
train_df = pd.read_csv("NSLKDD_top_10_TRAIN.csv")
test_df = pd.read_csv("NSLKDD_top_10_TEST.csv")
```

```python
train_df
```

```python
test_df
```

```python
train_x = train_df.drop(labels=['attack'],axis=1)
train_y = train_df['attack']
train_x
```

```python
test_x = test_df.drop(labels=['attack'],axis=1)
test_y = test_df['attack']
test_x
```

```python
from sklearn.neighbors import KNeighborsClassifier as knn
knn_clf = knn()
knn_clf.fit(train_x,train_y)
```

```python
from sklearn.metrics import classification_report,accuracy_score
ypred = knn_clf.predict(test_x)
print(classification_report(test_y,ypred))
```

```python
print(accuracy_score(test_y,ypred))
```
