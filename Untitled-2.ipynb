```python
from sklearn.preprocessing import LabelEncoder
cat_features =['protocol_type','service','flag','attack']
le = LabelEncoder()
for feature_name in cat_features:
    nslkdd_train_df_full[feature_name] = le.fit_transform(nslkdd_train_df_full[feature_name])
    print(f'Encoded classes for {feature_name} {le.classes_}')
    
    
```

```python
x = nslkdd_train_df_full.drop(labels=['attack'],axis=1).to_numpy()
y = nslkdd_train_df_full['attack'].to_numpy()
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
ordered_rank_features = SelectKBest(score_func=f_classif,k=5)
ordered_features = ordered_rank_features.fit(x,y)
ordered_features.scores_
```

```python
scores_df = pd.DataFrame(ordered_features.scores_,columns=['feature_score'])
ind_features = nslkdd_train_df_full.drop(labels=['attack'],axis=1)
features_df = pd.DataFrame(ind_features.columns,columns=['feature'])
```

```python
features_df.columns
```

```python
features_scores = pd.concat([features_df,scores_df],axis=1)
#features_scores.columns=['feature','feature_score']
features_scores
```

```python
top_10_features = features_scores.nlargest(10,'feature_score')
top_10_features
```

```python

new_df=nslkdd_train_df_full[top_10_features['feature']]
new_df
```

```python
a = new_df.to_numpy()
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(a,y)
```

# PRECISION
Precision = TP/(TP + FP)

Out of all the positive predictions made by the model how many of them are correct.

It is the accuracy of positive predictions

# RECALL
Recall = TP/(TP+FN)

Out of all the positive truths in the dataset how many of them were identified.
i.e; Out of all the 'U2R' predictions in the test dataset how many of them were identified.

Recall = TP/(TP+FN)

# FROM the above report it is clear that for the class 'U2R'

1. only 9% of the 'U2R' labels were identified in the predictions
2. out of all the identified labels of 'U2R' 100% of them are correctly labeled
